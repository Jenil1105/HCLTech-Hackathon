# *  **How Tables schema is interconnected?**

```python
from graphviz import Digraph

er = Digraph("Retail_ER_Diagram", format="png")
er.attr(rankdir="LR", size="10,8")

# Entities
er.node("Customer", "CUSTOMER\n(customer_id)")
er.node("SalesHeader", "STORE_SALES_HEADER\n(transaction_id)\ncustomer_id")
er.node("SalesLine", "STORE_SALES_LINE_ITEMS\n(line_item_id)\ntransaction_id\nproduct_id")
er.node("Product", "PRODUCTS\n(product_id)")
er.node("Store", "STORES\n(store_id)")
er.node("Promo", "PROMOTION_DETAILS\n(promotion_id)")
er.node("Loyalty", "LOYALTY_RULES\n(rule_id)")
er.node("Inventory", "STORE_INVENTORY\n(store_id, product_id)")

# Relationships
er.edge("Customer", "SalesHeader", label="places")
er.edge("SalesHeader", "SalesLine", label="contains")
er.edge("SalesLine", "Product", label="refers to")
er.edge("SalesLine", "Promo", label="applies")
er.edge("Customer", "Loyalty", label="earns points via")
er.edge("Store", "Inventory", label="has")
er.edge("Product", "Inventory", label="stocked as")

er
```

```python
from graphviz import Digraph

er = Digraph("Detailed_Retail_ER", format="png")
er.attr(rankdir="LR", size="14,10")

# Entities with attributes
er.node("Customer", """CUSTOMER_DETAILS
--------------------
PK customer_id
customer_name
email
loyalty_points""")

er.node("SalesHeader", """STORE_SALES_HEADER
--------------------
PK transaction_id
FK customer_id
FK store_id
transaction_date
total_amount""")

er.node("SalesLine", """STORE_SALES_LINE_ITEMS
--------------------
PK line_item_id
FK transaction_id
FK product_id
FK promotion_id
quantity
sales_amount""")

er.node("Product", """PRODUCTS
--------------------
PK product_id
product_name
category
price""")

er.node("Store", """STORES
--------------------
PK store_id
store_name
location""")

er.node("Promo", """PROMOTION_DETAILS
--------------------
PK promotion_id
promotion_type
discount_value
start_date
end_date""")

er.node("Loyalty", """LOYALTY_RULES
--------------------
PK rule_id
base_points_per_unit
bonus_threshold
bonus_points""")

er.node("Inventory", """STORE_INVENTORY
--------------------
PK inventory_id
FK store_id
FK product_id
inventory_date
stock_level""")

# Relationships
er.edge("Customer", "SalesHeader", label="places")
er.edge("SalesHeader", "SalesLine", label="contains")
er.edge("SalesLine", "Product", label="refers to")
er.edge("SalesLine", "Promo", label="applies")
er.edge("Store", "Inventory", label="maintains")
er.edge("Product", "Inventory", label="tracked in")

er
```
